```python
import numpy as np
import pandas as pd

import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc

import quandl
quandl.read_key()
```

```python
STR_FMT = '{0}\n{1}\n'
```

```python
# Get the data
df = quandl.get('EURONEXT/ABN', column_index=4)
print(STR_FMT.format('df.head()', df.head()))
print(STR_FMT.format('df.tail()', df.tail()))
df.plot()
```

```python
# Calculate and plot the percentage daily returns
daily_changes = df.pct_change(periods=1)
print(STR_FMT.format('df.head()', df.head()))
print(STR_FMT.format('daily_changes.head()', daily_changes.head()))
print(STR_FMT.format('daily_changes.describe()', daily_changes.describe()))
daily_changes.plot()
```

```python
# INCORRECT Calculate and plot the cumulative returns
df_cumprod = daily_changes.cumsum()
df_cumprod.plot()
(df / df['Last'][0] - 1).plot(ax=plt.gca())
```

```python
# Calculate and plot the cumulative returns
df_cumprod = (daily_changes + 1).cumprod() - 1
df_cumprod.plot()
(df / df['Last'][0] - 1).plot(ax=plt.gca())
```
